**(1) Data Description**
There are 2 files for the data: players.csv and sessions.csv
- The data was collected by a research group at UBC in Computer Science, led by Frank Wood. They set up a MineCraft serverLinks to an external site and recorded the players' actions as they navigated through the world.
- The dataset's are not in a tidy format. The "end_time" and "start_time" has to be separated into two columns: date and time.
  
**players.csv**
- 196 observations (196 rows)
- Mean of Age: 21.14
- Mean of played_hours: 5.85
- 7 variables
     - played_hours
       + A double variable (number that includes decimals) that reports the number of MineCraft hours played by
         each individual.
     - Age 
       + A double variable that report's the player's age.
     - experience 
       + A character variable with 5 categories. It indicates the level of MineCraft experience the individual player has. 
     - hashedEmail
       + A character variable that reports a player's hashed email address.
     - name
       + A character variable that reports a player's first name.
     - gender
       + A character variable that reports a player's gender.
     - subscribe
       + A logical vector variable (true or false) that tells you if a player is subscribed to a game-related
         newsletter.


    **sessions.csv**
- 1535 observations (1535 rows)
- 5 variables
     - original_start_time
       + A double variable that reports the "start_time" in UNIX time (milliseconds).
     - original_end_time
       + A double variable that reports the "end_time" in UNIX time (milliseconds).
     - hashedEmail
       + A character variable that reports a player's hashed email address.
     - start_time
       + A character variable that reports the start date and time of an individual playing MineCraft.
     - end_time
       + A character variable that reports the end date and time of an individual playing MineCraft.
            
          

**(2) Questions**

Broad Question:
What player characteristics and behaviours are most predictive of subscribing to a game-related newsletter, and how do these features differ between various player types?

Specific Question:
Can a player’s experience level and played hours predict whether they are subscribed to a game-related newsletter in the players.csv dataset?

- The players.csv dataset will address the specific question as it contains, subscribe (true/false), player experience (beginner to veteran) and played hours.
- Wrangling: subscribe will turn to factor while experience will turn to integer variables, TRUE/FALSE will turn to Yes/No, and experience will be ordered from level of player's experience from "Beginner" to "Veteran" (1 to 5).

**(3) Exploratory Data Analysis and Visualization:**

```r
library(tidyverse)
players <- read_csv("https://raw.githubusercontent.com/Ranaaelhosary/DSCI-100-Planning-Stage-Individual/refs/heads/main/players.csv")
sessions <- read_csv("https://raw.githubusercontent.com/Ranaaelhosary/DSCI-100-Planning-Stage-Individual/refs/heads/main/sessions.csv")
```

```r
#Converting data and assigning experience to numerical levels
players_mutated <- players |>
  mutate(subscribe = factor(subscribe, levels = c(FALSE, TRUE), labels = c("No", "Yes")),
    experience = as.integer(factor(experience, levels = c("Beginner", "Amateur", "Regular", "Pro", "Veteran"),
         labels = c(1, 2, 3, 4, 5) )) )         
```

```r
#Mean of quantitative variables
mean_age <- mean(players$Age, na.rm = TRUE)
mean_played_hours <- mean(players$played_hours, na.rm = TRUE)

mean_table <- tibble(Variable = c("Age", "played_hours"),
  Mean = c(mean_age, mean_played_hours))

mean_table
```

```r
#Number of subscribed players in each experience level
players_subscribed <- players_mutated |>
  filter(subscribe == "Yes") |>  
  group_by(experience) |>         
  summarise(count = n()) 

players_subscribed2 <- players |>
  filter(subscribe == "TRUE") |>  
  group_by(experience) |>         
  summarise(count = n()) 

players_subscribed
players_subscribed2
```

```r
#Scatter Plot of Data

options(repr.plot.width = 10, repr.plot.height = 10)
ggplot(players_mutated, aes(x = experience, y = played_hours, color = subscribe)) +
    geom_point(alpha = 0.6) +
    theme(text = element_text(size = 15)) +
          labs(title = "Newsletter Subscription by Player Experience and Played MineCraft Hours",
               x = "Experience Level",
               y = "Played MineCraft (in Hours)",
              color = "Subscribed to a Game-Related Newsletter")
```

* MineCraft player experience level and played hours that are subscribed to a game-related newletter.
* As played hours increases, the players are subscribed to the newsletter.
* The Amateur and Regular players play the most hours in the players.csv dataset.
* Limitation: Difficult to interpret how many individuals in each experience level are subscribed due to the overlapping
  points.

```r
#Bar Plot of Data
subscribed_counts <- players |>
  filter(subscribe == "TRUE")

options(repr.plot.width = 10, repr.plot.height = 10)
ggplot(subscribed_counts, aes(x = experience, fill = experience)) +
    geom_bar() +
    theme(text = element_text(size = 15)) +
          labs(title = "Newsletter Subscription by Player Experience",
            x = "Player Experience Level",
            y = "Number of Players Subscribed to a Game-Related Newsletter",
            fill = "Player Experience Level")
```

* Top MineCraft player experience level that is subscribed to a game-related newletter is the Amateur experience level (45), while the least is Pro (10) in the players.csv dataset. 
* Limitation: It only analyzes one player characteristic (experience) while there could be more factors at play.

**(4) Methods and Plan:**

* K-nearest neighbors (KNN) for Binary Classification on the Scatter plot.
  * Reason:
     + Player experience can be converted to numbers (Beginner = 1, ... , Veteran = 5), and hours played are numerical (allows
       KNN distance-based calculation).
     + The graph behaviour is non-linear
     + Predicting two classes (subscription Yes/No)
  * Assumptions:
     + Non-parametric: doesn't assume relationship between player characteristics and subscription.
     + Assumes players with similar experience and hours have the same subscription outcome.
  * Limitations:
     + Sensitive to outliers
     + Risk of overfitting (small K) or underfitting (large K)
 * Model:
     + K-fold cross-validation (5-10 fold) to choose the best K that causes smallest standard error and best accuracy.
  * Data Processing:
     + Convert experience to numerical values
     + Standardize the experience and hours played to contribute equally to distance calculations
     + Split data: Training Set = 80%, Testing Set = 20%
     + Apply cross validation on the Training Set
     + Fit KNN on Training Set with cross validation, evaluate on Testing Set
